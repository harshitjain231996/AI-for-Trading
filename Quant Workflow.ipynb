Fundamentally, quantitive trading is the process of using statistical analysis and modelling to predict market behavior. We use these predictions made to make trade with the goal of profit.

A good hypothesis is the first step to make predictions about future behavior. Coming up with a good hypothesis means being brutally specific. Example - Stocks that are discussed in the news are likely to go up - This statement is very vague and does not make sense since there is no specification of time or context while the discussion was going on about these stocks. An improved version would be - Stocks whose company name or ticker appears on the landing page of the wall street journal website, will increase in price by 1% one day following this appearance. Even though this statement is far from reality, but we can check against the ticker by entering the details.

Knowledge of market will be the deciding factor when it comes to designing a good hypothesis. If the hypothesis is making inconsistent predictions with real,observed data, then in that case the hypothesis is wrong

Hypothesis - An idea for a way to profit from trading. Generally, hypothesis which forms the basis of a trading strategy must survive several phases of increasingly rigorous testing. 

There is an initial research phase during which ideas are generated and validated in a basic way. The goal is to come up with promising ideas relatively quickly. In the research phase, we will focus on what set of positions to enter, on which assets and which time such that we have the potential to get positive future returns. Later on, full fleged trading strategy is to be designed where we can come up with more factors i.e. how much money to invest in each asset, under what conditions to exit positions, how to make trades and what risk and constraints to impose. The process of rigorously simulating the entire flow in an automated fashion using historical data is called backtesting.

## Flavors of trading strategies

1. Buying and selling a single asset.
2. Pairwise strategies - Testing against a pair of stocks which are related to each other. We trade based on their relative movements.
3. Cross-Sectional Strategies - Also known as equity statistical arbitrage or equity market neutral investing. It is a popular strategy and it involves comparing 100's to 1000's of stocks to determine which stocks lie in long and short portfolios with the goal of benefitting from transient market phenomena without being subject to overall market movements.The comparison is often based on price and volume or fundamental information. 
4. Alternative Data based Strategy - A new type of strategy which is primarily based on new type of data such as satellite imagery, social media, geolocation, consumer transaction data.

The 3rd and 4th strategies are most often used and are important.





## Anatomy of a Strategy

Stages of building a Cross-Sectional Strategy : 
    1. Data Selection
    2. Universal definition
    3. Alpha discovery
    4. Alpha Combination
    5. Portfolio construction
    6. Trading

We need to get the relevant data required to test it. Then we need to pair down a dataset with a subset that contains the stocks we want to potentially trade. It might be wise to exclude the stocks that have low trading volume, but at the same time we want to construct the portfolio in such a way that the stocks have similarities so that ranking them is reasonable. However since we will be looking to benefit from their movement relative to each other they shouldn't be too similar.

We also wanna limit our universe to the stocks to which our hypothesis logically applies. Example - The hypothesis discussed by Nguyen, Quoc in the geographic momentum, 2012 , said that a momentum effect is created for mnc's because investors do not pay attention to foreign market developments and so there is an opportunity to predict increases in these companies stock prices when markets in countries they are operating in improve. If we stick with this hypothesis, we only want to work with companies who actually have significant foreign market operations. 

Alpha : An expression that outputs a vector where each component is a value indicative of future returns for an individual stock. It is applied to the cross-section of our universe stock which outputs a vector of real numbers whose values are proportional to the size of the position one will take on each asset. When a successful alpha is discovered, the value in that vector for each stock is directly proportional to the rank of it's return at a future time across the universal stocks. For example - In a cross-sectional momentum strategy, we rank the stocks according to how much momentum they have as measured by a momentum indicator. According to their ranks, the stocks are either put in long portfolio or short portfolio at every time interval. So alpha is one type of trading signal. A trading signal is any numerical signal that informs a trade. In modern markets, it is very unlikely that a single value of alpha will provide us sufficiently consistent positive returns. Typically, several alphas will be combined together to generate an overall alpha that has better performance than the best individual alpha. Combining alpha that have diverse inputs and underlying hypothesis can lead to high quality combined alpha vector. For example - A price driven alpha such as momentum alpha may combine well with the company fundamental information alpha since the inputs are very different. Another method can be to translate alphas into features and use them as inputs to a machine learning classifier to capture the relationships between the alphas.

Portfolio Construction Stage - We want to start thinking about how our strategy will work in practice and this is the portfolio construction stage. The stage where we have to think about using combined alpha vector to generate and update an actual portfolio. 

There are different types of risk - 

1. Systematic Risk- The risks which are inherent to the entire market such as inflation, recession, interest rates, etc.
2. Sector specific Risk - Risk limited to a certain sector such as regulations, legislations, material costs etc.
3. Idiosyncratic Risk - Risk inherent to individual stocks. Example - Labor Strike, managerial change etc.

Different investment funds have different mandates driven by the goals of their investors. 

The portfolio can be designed to maximize different objectives. If one seeks to maximize expected returns but minimum variance, and if 2 portfolios have the same expected returns, the person should choose the one with lower variance.

The portfolio might be subject to additional requirements or constraints. For example - There might be a limit on what percentage of one's portfolio can be invested in any given stock, Some policies may dictate that one cannot hold stocks below a certain level of market capitalization, or one is only allowed to enter long positions to the government regulations in certain markets. These factors must be taken into account while designing the portfolio. 

In the trading stage, one makes actual trades in the market. After portfolio construction, one will have an ideal portfolio and a list of trades to make.
